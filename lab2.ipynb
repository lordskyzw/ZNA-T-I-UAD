```python
import pandas as pd

df = pd.read_csv('merged_dataset2.csv')

df.head()
```

PREDICTING DATA_USAGE

```python
from prophet import Prophet

df2 = df.drop(columns=['plan','total_messages_used', 'total_minutes'])
df2.head()
```

```python
specific_user_id = 1001 
filtered_df = df2[df2['user_id'] == specific_user_id]
filtered_df
```

```python
user_id10001 = filtered_df.drop(columns='user_id')
user_id10001
```

```python
user_id1k1 = user_id10001.rename(columns={'session_date': 'ds', 'mb_used': 'y'})
user_id1k1.head()
```

```python
m = Prophet()
m.fit(user_id1k1)
future = m.make_future_dataframe(periods=30)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(30)
```
